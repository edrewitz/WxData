```python
import requests

# 1. Parse the bucket and prefix from your URL
bucket = "ecmwf-open-data"
# Your path: 20260510/18z/ifs/0p25/scda/
prefix = "20260511/18z/ifs/0p25/scda/"

# 2. Construct the XML API URL
# This endpoint returns a 200 with content if files exist, 
# or an empty list if nothing matches the prefix.
api_url = f"https://storage.googleapis.com/{bucket}?prefix={prefix}"

def check_for_data(url):
    response = requests.get(url)
    
    if response.status_code != 200:
        print(f"❌ Error: Received status {response.status_code}")
        return False

    # In the XML API, if data exists, there will be <Contents> tags
    if "<Contents>" in response.text:
        print("✅ Data found at this location.")
        return True
    else:
        print("❌ No data available yet (Empty folder).")
        return False

check_for_data(api_url)
```
